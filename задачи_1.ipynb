**1 Последовательность чисел**

Даны два целых числа m и n. Напишите программу, которая выводит все целые числа от m до n включительно в порядке возрастания, если m < n, или в порядке убывания в противном случае.

```python
try:
    m = int(input("Введите первое число (m): "))
    n = int(input("Введите второе число (n): "))


    if m < n:

        print(f"Числа от {m} до {n} (возрастание):")
        for i in range(m, n + 1):
            print(i)
    else:

        print(f"Числа от {m} до {n} (убывание):")
        for i in range(m, n - 1, -1):
            print(i)

except ValueError:
    print("Ошибка: необходимо было ввести целые числа.")
```

## **2 Сумма чисел**

На вход программе подается натуральное число n. Напишите программу, которая подсчитывает сумму тех чисел от 1 до n (включительно), квадрат которых оканчивается на 2, 5 или 8.


```python
# 1. نطلب من المستخدم إدخال عدد طبيعي n
try:
    n = int(input("أدخل عددًا طبيعيًا n: "))

    # نتأكد أن العدد المدخل طبيعي (أكبر من 0)
    if n < 1:
        print("خطأ: يجب أن يكون n عددًا طبيعيًا (1 أو أكبر).")
    else:
        # 2. ننشئ متغيرًا لتخزين المجموع، ونبدأه بصفر
        total_sum = 0

        # 3. نبدأ حلقة تكرارية (loop) لكل عدد i من 1 إلى n
        # (استخدام n + 1 ضروري ليشمل n نفسها في الحلقة)
        for i in range(1, n + 1):

            # 4. نحسب مربع العدد i
            square = i * i

            # 5. نحصل على الرقم الأخير للمربع
            # (أسهل طريقة هي حساب باقي القسمة على 10)
            last_digit = square % 10

            # 6. نتحقق إذا كان الرقم الأخير هو 2، أو 5، أو 8
            # (يمكن كتابتها: if last_digit == 2 or last_digit == 5 or last_digit == 8)
            if last_digit in [2, 5, 8]:

                # 7. إذا تحقق الشرط، نضيف العدد الأصلي i (وليس مربعه) إلى المجموع
                total_sum += i

        # 8. بعد انتهاء الحلقة، نطبع المجموع النهائي
        print(f"مجموع الأعداد التي ينتهي مربعها بـ 2، 5، أو 8 هو: {total_sum}")

except ValueError:
    print("خطأ: الرجاء إدخال عدد صحيح سليم.")
```

**3 Наибольшие числа**

На вход программе подается натуральное число n (гарантируется, что n >= 2), а затем n различных натуральных чисел последовательности, каждое на отдельной строке. Напишите программу, которая выводит наибольшее и второе наибольшее число последовательности.

```python
try:
    # 1. نقرأ عدد الأرقام n
    n = int(input("أدخل عدد الأرقام (n >= 2): "))

    if n < 2:
        print("خطأ: n يجب أن تكون 2 أو أكبر.")
    else:
        # 2. نجهز المتغيرات بقيمة 0 (لأن الأعداد طبيعية >= 1)
        max1 = 0  # سيخزن أكبر عدد
        max2 = 0  # سيخزن ثاني أكبر عدد

        # 3. نبدأ حلقة لقراءة n أرقام
        for _ in range(n):
            num = int(input())

            # 4. نتحقق من الرقم الجديد
            if num > max1:
                # إذا كان الرقم الجديد هو الأكبر على الإطلاق
                max2 = max1  # الأكبر القديم يصبح هو الثاني
                max1 = num   # الرقم الجديد يصبح هو الأكبر
            elif num > max2:
                # إذا كان الرقم الجديد ليس أكبر من max1
                # ولكنه أكبر من max2
                max2 = num   # الرقم الجديد يصبح هو الثاني

        # 5. نطبع النتائج
        print(f"العدد الأكبر هو: {max1}")
        print(f"العدد الثاني الأكبر هو: {max2}")

except ValueError:
    print("خطأ: الرجاء إدخال أعداد صحيحة.")
```

**4 Только четные**

Напишите программу, которая считывает последовательность из 10 целых чисел (строка из цифр, разделенных пробелом) и определяет, является ли каждое из них чётным или нет. Нужно вывести "Все четные" или "Не все четные"

```python
# 1. نقرأ السطر ونقسمه إلى قائمة من النصوص
try:
    numbers_str = input("أدخل 10 أعداد مفصولة بمسافة: ").split()

    # 2. نستخدم دالة all() مع "generator expression"
    # int(num) % 2 == 0 يتحقق إذا كان العدد زوجيًا
    # all() ستعيد True فقط إذا كان الشرط صحيحًا لـ "كل" الأعداد

    if all(int(num) % 2 == 0 for num in numbers_str):
        print("Все четные")  # "كلها زوجية"
    else:
        print("Не все четные") # "ليست كلها زوجية"

except (ValueError, IndexError):
    print("خطأ: يرجى إدخال الأعداد بشكل صحيح.")
```

**5 Количество пятёрок**

На вход программе подается последовательность целых чисел от 1 до 5, каждое число на отдельной строке. Концом последовательности является любое неположительное число либо число, большее 5. Напишите программу, которая выводит количество пятерок.

```python
# 1. ننشئ عدادًا للرقم 5
count_fives = 0

print("أدخل الأعداد (من 1 إلى 5)، أدخل رقمًا آخر لإيقاف الإدخال:")

# 2. نبدأ حلقة لا نهائية
while True:
    try:
        # 3. نقرأ الرقم
        num = int(input())

        # 4. نتحقق من شرط التوقف
        if num <= 0 or num > 5:
            break  # نكسر الحلقة ونخرج

        # 5. نتحقق إذا كان الرقم هو 5
        if num == 5:
            count_fives += 1  # نزيد العداد

    except ValueError:
        print("الرجاء إدخال رقم صحيح.")
        continue # نواصل إلى الدورة التالية

# 6. نطبع النتيجة النهائية
print(f"عدد الخمسات التي تم إدخالها: {count_fives}")
```

...

```python
try:
    n = int(input("أدخل عدد الصفوف (n): "))

    # 1. نبدأ حلقة من 1 إلى n (شاملة)
    for i in range(1, n + 1):

        # 2. نحول الرقم i إلى نص (string)
        num_as_str = str(i)

        # 3. نضرب النص في i لتكراره i من المرات
        # مثال: "3" * 3 = "333"
        print(num_as_str * i)

except ValueError:
    print("خطأ: الرجاء إدخال عدد صحيح.")
```

**7 простые числа**

На вход программе подается два натуральных числа a и b (a < b). Напишите программу, которая находит все простые числа от a до b включительно.
**Примечание - число 1 простым не является.**

```python
try:
    a = int(input("أدخل الرقم الأول (a): "))
    b = int(input("أدخل الرقم الثاني (b): "))

    if a < 1: a = 1 # نتأكد أننا نبدأ من 1 على الأقل

    print(f"الأعداد الأولية بين {a} و {b} هي:")

    # 1. الحلقة الخارجية: تمر على كل رقم من a إلى b
    for num in range(a, b + 1):

        # 2. نتجاهل الرقم 1 فورًا (حسب الملاحظة)
        if num == 1:
            continue

        # 3. نفترض أن الرقم أولي مبدئيًا
        is_prime = True

        # 4. الحلقة الداخلية: لفحص القواسم
        # (نختبر فقط حتى الجذر التربيعي للرقم + 1)
        for i in range(2, int(num**0.5) + 1):

            # 5. إذا وجدنا قاسمًا...
            if num % i == 0:
                is_prime = False  # ... فهو ليس أوليًا
                break  # لا داعي لمواصلة فحص القواسم

        # 6. إذا بقي "is_prime" صحيحًا بعد الحلقة الداخلية
        # فهذا يعني أننا لم نجد أي قواسم، فالرقم أولي
        if is_prime:
            print(num)

except ValueError:
    print("خطأ: الرجاء إدخال أعداد صحيحة.")
```

**8 Нижний регистр**

На вход программе подается строка. Напишите программу, которая подсчитывает количество буквенных символов в нижнем регистре.

```python
text = input("أدخل النص: ")

count = 0

# 1. نمر على كل حرف في النص
for char in text:

    # 2. نتحقق إذا كان الحرف "حرفًا صغيرًا"
    if char.islower():
        count += 1

print(f"عدد الأحرف الصغيرة: {count}")

# (حل بايثون المختصر)
# count = sum(1 for char in text if char.islower())
# print(count)
```

**9 Количесвто цифр**

На вход программе подается строка текста. Напишите программу, которая подсчитывает количество цифр в данной строке.

```python
text = input("أدخل النص: ")

count = 0

# 1. نمر على كل حرف في النص
for char in text:

    # 2. نتحقق إذا كان الحرف "رقمًا"
    if char.isdigit():
        count += 1

print(f"عدد الأرقام في النص: {count}")

# (حل بايثون المختصر)
# count = sum(1 for char in text if char.isdigit())
# print(count)
```

**10 Первое и последнее вхождение**

На вход программе подаётся строка текста.

Если в этой строке буква «f» встречается только один раз, выведите её индекс.

Если она встречается два и более раза, выведите индексы её первого и последнего вхождения на одной строке, разделённые символом пробела.

Если буква «f» в данной строке не встречается, следует вывести «NO».

```python
text = input("أدخل النص: ")
char_to_find = 'f'

# 1. نحسب عدد مرات ظهور الحرف (هذا لا يتغير)
count = text.count(char_to_find)

# 2. نطبق الشروط
if count == 0:
    # لم يظهر أبدًا
    print("NO")
elif count == 1:
    # ظهر مرة واحدة
    # .find() تجد أول ظهور (بناءً على العد من 0)
    first_index = text.find(char_to_find)

    # نطبع الموقع + 1
    print(first_index + 1)
else:
    # ظهر مرتين أو أكثر
    first_index = text.find(char_to_find)  # أول ظهور (من 0)
    last_index = text.rfind(char_to_find) # آخر ظهور (من 0)

    # نطبع كلا الموقعين + 1
    print(f"{first_index + 1} {last_index + 1}")
```

**11 Автомобильный номер**

Корректный автомобильный номер (в России) имеет следующий формат:

<БУКВА><ЦИФРА><ЦИФРА><ЦИФРА><БУКВА><БУКВА>_<ЦИФРА><ЦИФРА>

или (когда у номера региона 3 цифры)

<БУКВА><ЦИФРА><ЦИФРА><ЦИФРА><БУКВА><БУКВА>_<ЦИФРА><ЦИФРА><ЦИФРА>

где <ЦИФРА> – это любая цифра, а <БУКВА> – это одна из букв кириллицы АВЕКМНОРСТУХ

Напишите программу, которая принимает на вход строку и проверяет, является ли эта строка корректным автомобильным номером. Программа должна вывести «YES» (без кавычек), если искусственный интеллект справился со своей задачей, или «NO» (без кавычек) в противном случае.

```python
def is_valid_plate(s):
    """
    تتحقق هذه الدالة مما إذا كان النص s يطابق تنسيق اللوحة.
    """
    allowed_letters = "ABCDEFG"

    # 1. نقسم النص عند علامة "_"
    parts = s.split(' ')

    # 2. يجب أن يكون هناك جزأين بالضبط
    if len(parts) != 2:
        return False

    part1, part2 = parts

    # --- 3. فحص الجزء الأول (مثل A123BC) ---

    # يجب أن يكون طوله 6
    if len(part1) != 6:
        return False

    # التحقق من كل حرف ورقمه
    # <حرف><رقم><رقم><رقم><حرف><حرف>
    if not (part1[0] in allowed_letters and
            part1[1].isdigit() and
            part1[2].isdigit() and
            part1[3].isdigit() and
            part1[4] in allowed_letters and
            part1[5] in allowed_letters):
        return False

    # --- 4. فحص الجزء الثاني (مثل 45 أو 456) ---

    # يجب أن يكون طوله 2 أو 3
    if not (len(part2) == 2 or len(part2) == 3):
        return False

    # يجب أن يتكون بالكامل من أرقام
    if not part2.isdigit():

        return False

    # 5. إذا نجحت كل الاختبارات
    return True

# --- البرنامج الرئيسي ---
plate_input = input("أدخل رقم اللوحة: ")

if is_valid_plate(plate_input):
    print("YES")
else:
    print("NO")
```

**12 Проверка никнейма**

Пользователь пытается создать никнейм для своего аккаунта. Правила для корректного никнейма следующие:

* никнейм должен начинаться с символа @
* никнейм должен содержать от 5 до 15 (включительно) символов (включая первый символ @)
* никнейм должен содержать **только** строчные буквы и цифры (помимо первого символа @)

Напишите программу, которая выводит «Correct» (без кавычек), если никнейм соответствует всем вышеприведенным правилам, или «Incorrect» (без кавычек) в противном случае.

Обратите внимание, что никнейму **необязательно** содержать строчные буквы и цифры одновременно, никнейм может содержать только строчные буквы или только цифры (помимо первого символа @).

```python
def is_valid_nickname(nick):
    """
    تتحقق هذه الدالة من صحة النيك نيم.
    """

    # 1. التحقق من البداية
    if not nick.startswith('@'):
        return False

    # 2. التحقق من الطول
    if not (5 <= len(nick) <= 15):
        return False

    # 3. التحقق من باقي الأحرف

    # s[1:] يعني "كل شيء ما عدا الحرف الأول"
    rest_of_nick = nick[1:]

    # نمر على كل حرف في "الباقي"
    for char in rest_of_nick:
        # إذا كان الحرف "ليس" حرفًا صغيرًا "و" "ليس" رقمًا
        if not (char.islower() or char.isdigit()):
            return False # وجدنا حرفًا غير صالح

    # 4. إذا نجحت كل الاختبارات
    return True

# --- البرنامج الرئيسي ---
nickname_input = input("أدخل النيك نيم: ")

if is_valid_nickname(nickname_input):
    print("Correct")
else:
    print("Incorrect")
```
